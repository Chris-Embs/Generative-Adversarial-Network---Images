```python
# CS 334: FINAL PROJECT | DEEP CONVOLUTIONAL GENERATIVE ADVERSARIAL NETWORK (DCGAN) w/ CIFAR-10
# Designed by Drew Lawless & Chris Embs; Finalized on 12/10/24

import os
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import time
from tensorflow.keras import layers, models
from skimage.metrics import structural_similarity as ssim
from sklearn.metrics import mean_squared_error

# prepares CIFAR-10 dataset for super-resolution and generator
def prepare_data():
    print("Retrieving CIFAR-10 dataset...")
    (train_images, _), (test_images, _) = tf.keras.datasets.cifar10.load_data()
    print(f"Dataset loaded! {len(train_images)} training images, {len(test_images)} testing images.")
    train_images = train_images.astype('float32') / 255.0
    test_images = test_images.astype('float32') / 255.0

    print("Reformatting images for lower resolution in generator...")
    train_lr = tf.image.resize(train_images, [16, 16])
    train_lr = tf.image.resize(train_lr, [32, 32])

    test_lr = tf.image.resize(test_images, [16, 16])
    test_lr = tf.image.resize(test_lr, [32, 32])
    print("CIFAR-10 preparation complete!")

    train_dataset = tf.data.Dataset.from_tensor_slices((train_lr, train_images)).shuffle(60000).batch(128)
    test_dataset = (test_lr, test_images)

    return train_dataset, test_dataset

# designs basic generator with super-resolution computation
def make_generator(input_shape=(32, 32, 3)):
    model = tf.keras.Sequential()
    model.add(layers.Input(shape=input_shape))
    model.add(layers.Conv2D(64, (3, 3), strides=(1, 1), padding='same'))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    for _ in range(4):
        model.add(layers.Conv2D(64, (3, 3), padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.LeakyReLU())

    model.add(layers.Conv2D(3, (3, 3), padding='same', activation='sigmoid'))

    return model

# design basic discriminator with super-resolution analysis
def make_discriminator(input_shape=(32, 32, 3)):
    model = tf.keras.Sequential()
    model.add(layers.Input(shape=input_shape))
    model.add(layers.Conv2D(64, (3, 3), strides=(2, 2), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))
    model.add(layers.Conv2D(128, (3, 3), strides=(2, 2), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))
    model.add(layers.Flatten())
    model.add(layers.Dense(1))

    return model

# used in loss plots/outputs
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)

# computes adversarial loss and MSE for total discriminator loss
def total_gen_loss(f_output, hr_images, g_images):
    adv_loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)(tf.ones_like(f_output), f_output)
    mse_loss = tf.reduce_mean(tf.square(hr_images - g_images))
    return adv_loss + mse_loss

# computes total discriminator loss
def total_disc_loss(r_output, f_output):
    r_loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)(tf.ones_like(r_output), r_output)
    f_loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)(tf.zeros_like(f_output), f_output)
    return r_loss + f_loss

# runs one training step on batch of images (391 steps per epoch)
@tf.function
def train_step(lr_images, hr_images, g, d, g_optimizer, d_optimizer):
    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        g_images = g(lr_images, training=True)
        r_output = d(hr_images, training=True)
        f_output = d(g_images, training=True)
        gen_loss = total_gen_loss(f_output, hr_images, g_images)
        disc_loss = total_disc_loss(r_output, f_output)

    g_gradients = gen_tape.gradient(gen_loss, g.trainable_variables)
    d_gradients = disc_tape.gradient(disc_loss, d.trainable_variables)
    g_optimizer.apply_gradients(zip(g_gradients, g.trainable_variables))
    d_optimizer.apply_gradients(zip(d_gradients, d.trainable_variables))

    return gen_loss, disc_loss

# computes MSE and SSIM scores for training and testing
def evaluation_metrics(g, dataset, batch_size=32):
    print("Calculating evaluation metrics...")
    mse_scores = []
    ssim_scores = []
    g.trainable = False

    for batch in dataset:
        lr_images, hr_images = batch
        g_images = g(lr_images, training=False)
        g_images = tf.clip_by_value(g_images * 255, 0, 255)
        hr_images = tf.clip_by_value(hr_images * 255, 0, 255)
        g_images = tf.cast(g_images, tf.uint8)
        hr_images = tf.cast(hr_images, tf.uint8)

        b_mse = np.mean([mean_squared_error(hr_images[i].numpy().flatten(), g_images[i].numpy().flatten())
                             for i in range(len(hr_images))])
        b_ssim = np.mean([ssim(hr_images[i].numpy(), g_images[i].numpy(),
                                    multichannel=True, win_size=3, data_range=255)
                              for i in range(len(hr_images))])

        mse_scores.append(b_mse)
        ssim_scores.append(b_ssim)

    avg_mse = np.mean(mse_scores)
    avg_ssim = np.mean(ssim_scores)

    return avg_mse, avg_ssim

# plots total g/d loss and evaluation metrics
def plot_eval_metrics(g_losses, d_losses, mse_scores, ssim_scores):
    print("Plotting total losses...")
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(g_losses, label="Total Generator Loss")
    plt.plot(d_losses, label="Total Discriminator Loss")
    plt.title("Generator & Discriminator Total Losses over Epochs")
    plt.xlabel("Epochs")
    plt.ylabel("Total Loss")
    plt.legend()
    print("Plotting MSE and SSIM...")
    plt.subplot(1, 2, 2)
    plt.plot(mse_scores, label="MSE")
    plt.plot(ssim_scores, label="SSIM")
    plt.title("MSE & SSIM Scores over Epochs")
    plt.xlabel("Epochs")
    plt.ylabel("Total Score")
    plt.legend()
    plt.tight_layout()
    plt.show()

#shows generated images and saves them every 5 epochs
def generate_save_images(g, epoch, test_images, save_dir='/content/generated_images'):
    print(f"Generating and saving images for Epoch {epoch}...")
    g_images = g(test_images, training=False)
    g_images = tf.clip_by_value(g_images, 0, 1)
    os.makedirs(save_dir, exist_ok=True)

    for i, img in enumerate(g_images):
        img = (img.numpy() * 255).astype("uint8")
        g_image_path = os.path.join(save_dir, f"epoch_{epoch:03d}_image_{i + 1}.png")
        plt.imsave(g_image_path, img)
    print("Generated images saved!")

    fig, axes = plt.subplots(1, len(test_images), figsize=(15, 5))
    for i, img in enumerate(g_images[:len(test_images)]):
        axes[i].imshow(img)
        axes[i].axis('off')
    plt.tight_layout()
    plt.show()

# performs actual training/testing of g/d
def train(g, d, dataset, ep, test_images):
    print("Beginning training...\n")
    g_optimizer = tf.keras.optimizers.Adam(1e-4)
    d_optimizer = tf.keras.optimizers.Adam(1e-4)
    gl_record = []
    dl_record = []
    mse_record = []
    ssim_record = []
    example_test_images = test_images[:5]

    for epoch in range(ep):
        start = time.time()
        print(f"Epoch {epoch + 1}/{ep} started...")

        for step, (lr_images, hr_images) in enumerate(dataset):
            gen_loss, dis_loss = train_step(lr_images, hr_images, g, d, g_optimizer, d_optimizer)
            if step % 5 == 0 or step == 391:
                print(f"    Step {step}: Total Generator Loss: {gen_loss:.4f}, Total Discriminator Loss: {dis_loss:.4f}")

        gl_record.append(gen_loss)
        dl_record.append(dis_loss)
        avg_mse, avg_ssim = evaluation_metrics(g, dataset)
        mse_record.append(avg_mse)
        ssim_record.append(avg_ssim)
        print(f"Epoch {epoch + 1}: MSE: {avg_mse}, SSIM: {avg_ssim}")
        print(f"Epoch {epoch + 1} runtime: {time.time() - start:.2f} sec.")

        if (epoch + 1) % 1 == 0:
            with tf.device('/CPU:0'):
                generate_save_images(g, epoch + 1, example_test_images)

    print("All training complete!")
    plot_eval_metrics(gl_record, dl_record, mse_record, ssim_record)

#runs data processing, model creation, training, testing, and plotting
g = make_generator()
d = make_discriminator()
train_dataset, (test_lr, test_images) = prepare_data()
train(g, d, train_dataset, ep=100, test_images=test_images)
```
